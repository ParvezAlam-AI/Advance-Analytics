Dropping irrelevant columns

```python
df = df.drop(['Engine Fuel Type','Market Category','Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'],axis=1)
df.head(5)
```

renaming the columns

```python
df = df.rename(columns={"Engine HP":"HP", "Engine Cylinders":"Cylinders", "Transmission Type":"Transmission","Driven_Wheels": "Drive Mode","highway MPG":"MPG-H","chiy mpg":"MPG-H","MSRP":"Price"})
df.head(5)
```

dropping duplicate rows

```python
df.shape
```

```python
dup_rows_df = df[df.duplicated()]
print("number of duplicate rows: ",dup_rows_df)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
print(df.shape)
print(df.count())
```

dropping the missing or null values

```python
print(df.isnull().sum())
```

This is the reason in the above step while counting both Cylinders and Horsepower (HP) had 10856 and 10895 over 10925 rows.

```python
df = df.dropna()  #dropping the missing value
df.count()
```

Now we have removed all the rows which contain the Null or N/A values (Cylinders and Horsepower (HP)).


```python
print(df.isna().sum())
```

# Plot different features against one anatoher (scatter), against frequency (histogram)

```python
#Histogram
df.Make.value_counts().nlargest(40).plot(kind = 'bar',figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
df.HP.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

### Heat Maps

Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps. In the below heat map we know that the price feature depends mainly on the Engine Size, Horsepower, and Cylinders.

```python
# plt.figure(figsize=(10,5))
# c= df.corr()
# sns.heatmap(c,cmap="BrBG",annot=True)
# c

# # here we are getting string to float error in one of the columns. so we are not taking numeric columns to generate heatmap plot
```

```python
df_numeric = df.select_dtypes(include = [float,int])
c = df_numeric.corr()
```

```python
chart_size = (10,5)
plt.figure(figsize = chart_size)
sns.heatmap(c,cmap="BrBG",annot=True)
plt.show()
```

## scatterplot


We generally use scatter plots to find the correlation between two variables. Here the scatter plots are plotted between Horsepower and Price and we can see the plot below. With the plot given below, we can easily draw a trend line. These features provide a good scattering of points.


```python
fig, ax = plt.subplots(figsize = chart_size)
ax.scatter(df['HP'],df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
